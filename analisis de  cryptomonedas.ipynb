```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests as req
import numpy as np
```

```python
url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=100&page=1&sparkline=false"

response = req.get(url)

data = response.json()

df = pd.DataFrame(data)
df.head()
print(df.head())
print(df.columns)
```

```python
nulos = df.isnull().sum()
print(nulos[nulos > 0])
```

```python
# --- 2. CONFIGURACIÃ“N ESTÃ‰TICA DE DATOS---
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 15)) 
plt.suptitle('AnÃ¡lisis de Mercado Cripto (Top 100)', fontsize=24, weight='bold')
```

```python
# --- GRÃFICO A: HISTOGRAMA (DistribuciÃ³n) ---
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='current_price', kde=True, color='skyblue', log_scale=True)
plt.title('A) DistribuciÃ³n de Precios (Escala Log)', fontsize=16)
plt.xlabel('Precio Actual (USD)')
plt.ylabel('Frecuencia')
```

```python
# --- GRÃFICO B: BOXPLOT (Outliers) ---
plt.subplot(2, 2, 2)
sns.boxplot(x=df['current_price'], color='lightgreen')
plt.xscale('log')
plt.title('B) DetecciÃ³n de Outliers de Precio', fontsize=16)
plt.xlabel('Precio Actual (USD)')
```

```python
# --- GRÃFICO C: BARRAS (Top 10 MÃ¡s Caras) ---
plt.subplot(2, 2, 3)
top_10_expensive = df.nlargest(10, 'current_price')
sns.barplot(data=top_10_expensive, x='current_price', y='name', palette='viridis')
plt.title('C) Las 10 Criptos con Mayor Precio Unitario', fontsize=16)
plt.xlabel('Precio (USD)')
plt.ylabel('')
```

```python
# --- GRÃFICO DispersiÃ³n ---
plt.subplot(2, 2, 4)
sns.scatterplot(data=df, x='market_cap', y='current_price', 
                hue='price_change_percentage_24h', palette='coolwarm', size='total_volume', sizes=(20, 200))
plt.xscale('log')
plt.yscale('log')
plt.title('D) Precio vs. CapitalizaciÃ³n de Mercado', fontsize=16)
plt.xlabel('Market Cap ')
plt.ylabel('Precio ')
plt.legend(title='Cambio 24h %', bbox_to_anchor=(1.05, 1), loc='upper left')
```
