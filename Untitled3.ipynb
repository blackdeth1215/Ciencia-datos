Estos dispositivos registran temperatura, humedad y estado del sensor cada 30 minutos, enviando la información a un servidor interno para su análisis y control de condiciones ambientales

```python
archivo = pd.read_csv('iot_sensor_corrupted_dataset.csv', na_values=['??', '?', 'N/A', 'NaN'])
archivo.columns = archivo.columns.str.strip().str.replace('#', '')
archivo = archivo.loc[:, ~archivo.columns.str.contains('^Unnamed')]
df_raw = pd.DataFrame(archivo)
print(f"Filas antes de limpiar: {df_raw.shape[0]}")
```

#Limpiesa de datos y con informacion incompleta

```python
nulos = df_raw.isnull().sum()
print("datos nulos: \n"+nulos[nulos > 0].to_string())
```

```python
if not df_raw.empty:

    duplicados = df_raw.duplicated().sum()
    if duplicados > 0:
        print(f"Se encontraron {duplicados} registros duplicados. Eliminando redundancia...")
        df_clean = df_raw.drop_duplicates()
    else:
        df_clean = df_raw.copy()
    df_clean['timestamp'] = pd.to_datetime(df_clean['timestamp'], errors='coerce')
    fechas_invalidas = df_clean['timestamp'].isnull().sum()
    fecha_limite = pd.Timestamp.now() + pd.Timedelta(days=1) 
    fechas_futuras = df_clean[df_clean['timestamp'] > fecha_limite].shape[0]
    print(f"- Fechas con formato inválido (corruptas): {fechas_invalidas}")
    print(f"- Fechas futuras (imposibles): {fechas_futuras}")

    df_clean = df_clean.dropna(subset=['timestamp'])
    df_clean = df_clean[df_clean['timestamp'] <= fecha_limite]

    df_clean['temperature_C'] = pd.to_numeric(df_clean['temperature_C'], errors='coerce')
    df_clean['humidity_%'] = pd.to_numeric(df_clean['humidity_%'], errors='coerce')

    temp_anomalias = df_clean[(df_clean['temperature_C'] > 50) | (df_clean['temperature_C'] < -10)].shape[0]
    hum_anomalias = df_clean[(df_clean['humidity_%'] > 100) | (df_clean['humidity_%'] < 0)].shape[0]

    print("\n[!] Fallas de Hardware Detectadas:")
    print(f"    - Temperaturas fuera de rango (>50°C o <-10°C): {temp_anomalias}")
    print(f"    - Humedad imposible (>100% o <0%): {hum_anomalias}")

    df_clean.loc[(df_clean['temperature_C'] > 50) | (df_clean['temperature_C'] < -10), 'temperature_C'] = np.nan
    df_clean.loc[(df_clean['humidity_%'] > 100) | (df_clean['humidity_%'] < 0), 'humidity_%'] = np.nan

    print("\n--- Rellenando datos faltantes (Interpolación) ---")
    
    df_clean = df_clean.sort_values('timestamp') 
    
    df_clean['temperature_C'] = df_clean['temperature_C'].interpolate(method='linear')
    df_clean['humidity_%'] = df_clean['humidity_%'].interpolate(method='linear')
    
    df_clean['temperature_C'] = df_clean['temperature_C'].ffill()
    df_clean['status'] = df_clean['status'].fillna('Unknown')

    print("Datos saneados y listos para análisis.")
    print(df_raw)
```
